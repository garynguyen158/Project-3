```
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
import datetime
import numpy as np
from datetime import timedelta
from datetime import datetime
import matplotlib.dates as md
import time
import folium
import plotly.graph_objects as go
%matplotlib inline
```

```
data = 'https://raw.githubusercontent.com/garynguyen158/Project-3/master/DATA3.csv'
data = pd.read_csv(data)
```

```
T_raw = data.loc[:,'time']
Lat = data.loc[:,'latitude']
Lon = data.loc[:,'longitude']
Mag = data.loc[:,'mag']
```

```
Time = []
new_format = "%Y-%m-%d %H:%M:%S"
for i in range(len(T_raw)):
  datecheck = datetime.strptime(T_raw[i],"%Y-%m-%dT%H:%M:%S.%fZ").replace(microsecond=0)
  datecheck.strftime(new_format)
  Time.append(datecheck)
```

```
fig = go.Figure()
fig.add_trace(go.Scatter(x=Time, y=Mag,line_color='deepskyblue'))
fig.update_traces(
    marker={"size": 8},
    mode="markers",)
fig.update_layout(
    title="Magnitude of all Earthquake From 2010-Jan-01",
    xaxis_title="Timeline",
    yaxis_title="Magnitude of Earthqakes",
    font=dict(
        family="Courier New, monospace",
        size=18,
        color="black"
    )
)
fig.show()
```

```
folium_map = folium.Map(location=[46.244723, -125.423272],
                            zoom_start=5,
                            width='30%', height = '70%')
for i in range(len(Mag)):
    folium.CircleMarker(location=(Lat[i],Lon[i]),
                        radius=Mag[i],
                        fill=True).add_to(folium_map)
folium_map
```
